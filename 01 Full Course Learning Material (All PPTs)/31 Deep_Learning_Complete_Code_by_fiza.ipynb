## 2. Section 1: Single Neuron from Scratch

Let's implement a single neuron exactly as described in the documentation:

**z = (w₁·x₁ + w₂·x₂ + ... + wₙ·xₙ) + b**

**a = f(z)**

```python
class Neuron:
    """
    A single artificial neuron.
    
    Parameters:
    -----------
    n_inputs : int
        Number of input features
    activation : str
        Activation function ('relu', 'sigmoid', 'tanh', or 'linear')
    """
    
    def __init__(self, n_inputs, activation='relu'):
        # Initialize weights randomly (small values)
        self.weights = np.random.randn(n_inputs) * 0.1
        # Initialize bias to zero
        self.bias = 0.0
        self.activation = activation
        
    def forward(self, x):
        """
        Forward pass through the neuron.
        
        Parameters:
        -----------
        x : array-like
            Input values
            
        Returns:
        --------
        activation : float
            Output after activation function
        """
        # Step 1: Weighted sum
        z = np.dot(self.weights, x) + self.bias
        
        # Step 2: Activation
        if self.activation == 'relu':
            a = np.maximum(0, z)
        elif self.activation == 'sigmoid':
            a = 1 / (1 + np.exp(-z))
        elif self.activation == 'tanh':
            a = np.tanh(z)
        else:  # linear
            a = z
            
        return a

print("✓ Neuron class defined")
```

### Example 1: Recreate the Worked Example from Documentation

**Given:**
- Inputs: x₁ = 1.0, x₂ = 2.0
- Weights: w₁ = 0.5, w₂ = 0.3
- Bias: b = 0.1
- Activation: ReLU

```python
# Create neuron with specific weights
neuron = Neuron(n_inputs=2, activation='relu')
neuron.weights = np.array([0.5, 0.3])
neuron.bias = 0.1

# Input
x = np.array([1.0, 2.0])

# Compute output
output = neuron.forward(x)

print("=== Neuron Computation ===")
print(f"Inputs: x₁ = {x[0]}, x₂ = {x[1]}")
print(f"Weights: w₁ = {neuron.weights[0]}, w₂ = {neuron.weights[1]}")
print(f"Bias: b = {neuron.bias}")
print()
print("Step-by-step:")
print(f"  w₁·x₁ = {neuron.weights[0]} × {x[0]} = {neuron.weights[0] * x[0]}")
print(f"  w₂·x₂ = {neuron.weights[1]} × {x[1]} = {neuron.weights[1] * x[1]}")
print(f"  Sum = {neuron.weights[0] * x[0]} + {neuron.weights[1] * x[1]} = {neuron.weights[0] * x[0] + neuron.weights[1] * x[1]}")
z = neuron.weights[0] * x[0] + neuron.weights[1] * x[1] + neuron.bias
print(f"  z = sum + bias = {z - neuron.bias} + {neuron.bias} = {z}")
print(f"  ReLU(z) = max(0, {z}) = {output}")
print()
print(f"Final Output: {output}")
print("\n✓ This matches the documentation example!")
```

## 3. Section 2: Activation Functions

Let's implement and visualize all common activation functions:

```python
def relu(z):
    """ReLU: Rectified Linear Unit"""
    return np.maximum(0, z)

def sigmoid(z):
    """Sigmoid activation"""
    return 1 / (1 + np.exp(-z))

def tanh(z):
    """Hyperbolic tangent"""
    return np.tanh(z)

def leaky_relu(z, alpha=0.01):
    """Leaky ReLU"""
    return np.where(z > 0, z, alpha * z)

def softmax(z):
    """Softmax (for multi-class classification)"""
    exp_z = np.exp(z - np.max(z))  # Subtract max for numerical stability
    return exp_z / np.sum(exp_z)

print("✓ Activation functions defined")
```

```python
# Visualize activation functions
z = np.linspace(-5, 5, 200)

plt.figure(figsize=(15, 10))

# ReLU
plt.subplot(2, 3, 1)
plt.plot(z, relu(z), linewidth=2)
plt.title('ReLU: max(0, z)', fontsize=12, fontweight='bold')
plt.xlabel('z (pre-activation)')
plt.ylabel('activation')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

# Sigmoid
plt.subplot(2, 3, 2)
plt.plot(z, sigmoid(z), linewidth=2, color='orange')
plt.title('Sigmoid: 1/(1+e⁻ᶻ)', fontsize=12, fontweight='bold')
plt.xlabel('z (pre-activation)')
plt.ylabel('activation')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axhline(y=1, color='k', linewidth=0.5, linestyle='--', alpha=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

# Tanh
plt.subplot(2, 3, 3)
plt.plot(z, tanh(z), linewidth=2, color='green')
plt.title('Tanh', fontsize=12, fontweight='bold')
plt.xlabel('z (pre-activation)')
plt.ylabel('activation')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axhline(y=1, color='k', linewidth=0.5, linestyle='--', alpha=0.5)
plt.axhline(y=-1, color='k', linewidth=0.5, linestyle='--', alpha=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

# Leaky ReLU
plt.subplot(2, 3, 4)
plt.plot(z, leaky_relu(z), linewidth=2, color='red')
plt.title('Leaky ReLU: max(αz, z)', fontsize=12, fontweight='bold')
plt.xlabel('z (pre-activation)')
plt.ylabel('activation')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

# Linear
plt.subplot(2, 3, 5)
plt.plot(z, z, linewidth=2, color='purple')
plt.title('Linear: f(z) = z', fontsize=12, fontweight='bold')
plt.xlabel('z (pre-activation)')
plt.ylabel('activation')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

# Softmax (example with 3 classes)
plt.subplot(2, 3, 6)
z_softmax = np.array([-1, 0, 3])  # Example logits for 3 classes
probs = softmax(z_softmax)
plt.bar(['Class 1', 'Class 2', 'Class 3'], probs, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Softmax Example\n(converts logits to probabilities)', fontsize=12, fontweight='bold')
plt.ylabel('Probability')
plt.ylim(0, 1)
for i, prob in enumerate(probs):
    plt.text(i, prob + 0.02, f'{prob:.3f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("✓ Activation functions visualized")
```

## 5. Section 4: Loss Functions

Let's implement all common loss functions:

```python
def mean_squared_error(y_true, y_pred):
    """
    Mean Squared Error (MSE) - for regression
    
    L = (1/n) Σ (y_true - y_pred)²
    """
    return np.mean((y_true - y_pred) ** 2)

def mean_absolute_error(y_true, y_pred):
    """
    Mean Absolute Error (MAE) - for regression
    
    L = (1/n) Σ |y_true - y_pred|
    """
    return np.mean(np.abs(y_true - y_pred))

def binary_cross_entropy(y_true, y_pred):
    """
    Binary Cross-Entropy - for binary classification
    
    L = -(1/n) Σ [y_true·log(y_pred) + (1-y_true)·log(1-y_pred)]
    """
    # Clip predictions to prevent log(0)
    y_pred = np.clip(y_pred, 1e-7, 1 - 1e-7)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

def categorical_cross_entropy(y_true, y_pred):
    """
    Categorical Cross-Entropy - for multi-class classification
    
    L = -(1/n) Σ Σ y_true_c · log(y_pred_c)
    
    y_true should be one-hot encoded
    """
    # Clip predictions to prevent log(0)
    y_pred = np.clip(y_pred, 1e-7, 1 - 1e-7)
    return -np.mean(np.sum(y_true * np.log(y_pred), axis=1))

print("✓ Loss functions implemented")
```

```python
# Example: Compare loss functions
print("=== Loss Function Examples ===")
print()

# Regression example
y_true_reg = np.array([3.0, 2.5, 4.0, 3.5])
y_pred_reg = np.array([2.8, 2.6, 3.8, 3.7])

print("Regression Example:")
print(f"  True values:      {y_true_reg}")
print(f"  Predicted values: {y_pred_reg}")
print(f"  MSE:  {mean_squared_error(y_true_reg, y_pred_reg):.4f}")
print(f"  MAE:  {mean_absolute_error(y_true_reg, y_pred_reg):.4f}")
print()

# Binary classification example
y_true_bin = np.array([1, 0, 1, 1, 0])
y_pred_bin = np.array([0.9, 0.1, 0.8, 0.7, 0.2])

print("Binary Classification Example:")
print(f"  True labels:      {y_true_bin}")
print(f"  Predicted probs:  {y_pred_bin}")
print(f"  Binary Cross-Entropy: {binary_cross_entropy(y_true_bin, y_pred_bin):.4f}")
print()

# Multi-class example
y_true_cat = np.array([
    [1, 0, 0],  # Class 0
    [0, 1, 0],  # Class 1
    [0, 0, 1]   # Class 2
])
y_pred_cat = np.array([
    [0.8, 0.1, 0.1],
    [0.1, 0.7, 0.2],
    [0.05, 0.05, 0.9]
])

print("Multi-class Classification Example:")
print(f"  True labels (one-hot):\n{y_true_cat}")
print(f"  Predicted probabilities:\n{y_pred_cat}")
print(f"  Categorical Cross-Entropy: {categorical_cross_entropy(y_true_cat, y_pred_cat):.4f}")

print("\n✓ Loss functions demonstrated")
```

## 6. Section 5: Backpropagation from Scratch

Let's implement backpropagation manually to understand the mechanics:

```python
def relu_derivative(z):
    """Derivative of ReLU"""
    return (z > 0).astype(float)

def sigmoid_derivative(a):
    """Derivative of sigmoid (given activation a, not z)"""
    return a * (1 - a)

class SimpleNetwork:
    """
    A simple 2-layer neural network implemented from scratch.
    Input → Hidden (ReLU) → Output (Linear)
    """
    
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.01):
        # Initialize weights
        self.W1 = np.random.randn(hidden_size, input_size) * 0.1
        self.b1 = np.zeros(hidden_size)
        self.W2 = np.random.randn(output_size, hidden_size) * 0.1
        self.b2 = np.zeros(output_size)
        self.lr = learning_rate
        
    def forward(self, x):
        """Forward propagation"""
        # Hidden layer
        self.z1 = np.dot(self.W1, x) + self.b1
        self.a1 = relu(self.z1)
        
        # Output layer
        self.z2 = np.dot(self.W2, self.a1) + self.b2
        self.a2 = self.z2  # Linear activation
        
        return self.a2
    
    def backward(self, x, y_true, y_pred):
        """
        Backpropagation - compute gradients and update weights
        """
        # Output layer gradients
        # For MSE loss: dL/dy_pred = 2(y_pred - y_true)
        dL_da2 = 2 * (y_pred - y_true)
        
        # Since output activation is linear: dz2/da2 = 1
        dL_dz2 = dL_da2
        
        # Gradients for W2 and b2
        dL_dW2 = np.outer(dL_dz2, self.a1)  # Outer product
        dL_db2 = dL_dz2
        
        # Hidden layer gradients
        dL_da1 = np.dot(self.W2.T, dL_dz2)
        dL_dz1 = dL_da1 * relu_derivative(self.z1)
        
        # Gradients for W1 and b1
        dL_dW1 = np.outer(dL_dz1, x)
        dL_db1 = dL_dz1
        
        # Update weights (gradient descent)
        self.W2 -= self.lr * dL_dW2
        self.b2 -= self.lr * dL_db2
        self.W1 -= self.lr * dL_dW1
        self.b1 -= self.lr * dL_db1
    
    def train_step(self, x, y_true):
        """Single training step"""
        # Forward pass
        y_pred = self.forward(x)
        
        # Compute loss
        loss = mean_squared_error(y_true, y_pred)
        
        # Backward pass
        self.backward(x, y_true, y_pred)
        
        return loss

print("✓ SimpleNetwork class implemented with manual backpropagation")
```

```python
# Train the network on a simple dataset
print("=== Training Network from Scratch ===")
print()

# Create simple dataset: y = 2*x1 + 3*x2
np.random.seed(42)
n_samples = 100
X_train = np.random.randn(n_samples, 2)
y_train = 2 * X_train[:, 0] + 3 * X_train[:, 1] + np.random.randn(n_samples) * 0.1

# Create and train network
network = SimpleNetwork(input_size=2, hidden_size=5, output_size=1, learning_rate=0.01)

losses = []
epochs = 1000

for epoch in range(epochs):
    epoch_loss = 0
    for i in range(len(X_train)):
        loss = network.train_step(X_train[i], y_train[i:i+1])
        epoch_loss += loss
    
    avg_loss = epoch_loss / len(X_train)
    losses.append(avg_loss)
    
    if (epoch + 1) % 200 == 0:
        print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.6f}")

# Plot training loss
plt.figure(figsize=(10, 6))
plt.plot(losses, linewidth=2)
plt.title('Training Loss (Manual Backpropagation)', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.grid(True, alpha=0.3)
plt.show()

print("\n✓ Network trained successfully with manual backpropagation!")
```

## 7. Section 6: Complete Neural Network (NumPy)

Let's build a more complete implementation with batch processing:

```python
class NeuralNetwork:
    """
    A complete neural network implementation in NumPy.
    Supports:
    - Multiple layers
    - Batch processing
    - Different activation functions
    - Mini-batch gradient descent
    """
    
    def __init__(self, layer_sizes, learning_rate=0.01):
        """
        Parameters:
        -----------
        layer_sizes : list
            List of layer sizes [input, hidden1, hidden2, ..., output]
        learning_rate : float
            Learning rate for gradient descent
        """
        self.layer_sizes = layer_sizes
        self.lr = learning_rate
        self.n_layers = len(layer_sizes) - 1
        
        # Initialize weights and biases
        self.weights = []
        self.biases = []
        
        for i in range(self.n_layers):
            # He initialization for weights
            w = np.random.randn(layer_sizes[i+1], layer_sizes[i]) * np.sqrt(2.0 / layer_sizes[i])
            b = np.zeros((layer_sizes[i+1], 1))
            self.weights.append(w)
            self.biases.append(b)
    
    def forward(self, X):
        """
        Forward propagation
        
        Parameters:
        -----------
        X : array, shape (n_features, n_samples)
            Input data
            
        Returns:
        --------
        activations : list
            Activations for each layer
        """
        self.activations = [X]
        self.z_values = []
        
        A = X
        
        for i in range(self.n_layers):
            Z = np.dot(self.weights[i], A) + self.biases[i]
            
            # Apply activation (ReLU for hidden, linear for output)
            if i < self.n_layers - 1:
                A = relu(Z)
            else:
                A = Z  # Linear for output
            
            self.z_values.append(Z)
            self.activations.append(A)
        
        return A
    
    def backward(self, X, y):
        """
        Backpropagation
        
        Parameters:
        -----------
        X : array, shape (n_features, n_samples)
            Input data
        y : array, shape (n_outputs, n_samples)
            True labels
        """
        m = X.shape[1]  # Number of samples
        
        # Output layer gradient
        dZ = self.activations[-1] - y
        
        # Backpropagate through layers
        for i in range(self.n_layers - 1, -1, -1):
            # Compute gradients
            dW = (1/m) * np.dot(dZ, self.activations[i].T)
            db = (1/m) * np.sum(dZ, axis=1, keepdims=True)
            
            if i > 0:
                # Gradient for next layer
                dA = np.dot(self.weights[i].T, dZ)
                dZ = dA * relu_derivative(self.z_values[i-1])
            
            # Update weights
            self.weights[i] -= self.lr * dW
            self.biases[i] -= self.lr * db
    
    def train(self, X, y, epochs=1000, batch_size=32, verbose=True):
        """
        Train the network
        
        Parameters:
        -----------
        X : array, shape (n_samples, n_features)
            Training data
        y : array, shape (n_samples,) or (n_samples, n_outputs)
            Training labels
        epochs : int
            Number of training epochs
        batch_size : int
            Batch size for mini-batch gradient descent
        verbose : bool
            Print training progress
        """
        # Reshape data
        X = X.T  # (n_features, n_samples)
        y = y.reshape(-1, X.shape[1])  # (n_outputs, n_samples)
        
        n_samples = X.shape[1]
        losses = []
        
        for epoch in range(epochs):
            # Shuffle data
            indices = np.random.permutation(n_samples)
            X_shuffled = X[:, indices]
            y_shuffled = y[:, indices]
            
            epoch_loss = 0
            
            # Mini-batch training
            for i in range(0, n_samples, batch_size):
                X_batch = X_shuffled[:, i:i+batch_size]
                y_batch = y_shuffled[:, i:i+batch_size]
                
                # Forward and backward
                y_pred = self.forward(X_batch)
                self.backward(X_batch, y_batch)
                
                # Compute loss
                batch_loss = mean_squared_error(y_batch, y_pred)
                epoch_loss += batch_loss
            
            avg_loss = epoch_loss / (n_samples // batch_size + 1)
            losses.append(avg_loss)
            
            if verbose and (epoch + 1) % 100 == 0:
                print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.6f}")
        
        return losses
    
    def predict(self, X):
        """Make predictions"""
        X = X.T
        return self.forward(X).T

print("✓ Complete NeuralNetwork class implemented")
```

```python
# Test the complete network on a regression problem
print("=== Training Complete Neural Network ===")
print()

# Generate dataset
np.random.seed(42)
n_samples = 500
X = np.random.randn(n_samples, 3)  # 3 features
y = 2*X[:, 0] + 3*X[:, 1] - X[:, 2] + np.random.randn(n_samples) * 0.5

# Create network: 3 → 10 → 5 → 1
network = NeuralNetwork(layer_sizes=[3, 10, 5, 1], learning_rate=0.01)

print(f"Network architecture: {network.layer_sizes}")
print()

# Train
losses = network.train(X, y, epochs=500, batch_size=32, verbose=True)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(losses, linewidth=2)
plt.title('Training Loss (Complete NumPy Network)', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.grid(True, alpha=0.3)
plt.show()

print("\n✓ Complete network trained successfully!")
```

## 9. Section 8: PyTorch Implementations

Now the same examples in PyTorch:

```python
# Simple regression model in PyTorch
print("=== PyTorch Implementation ===")
print()

# Generate dataset
np.random.seed(42)
X_train_np = np.random.randn(500, 2).astype(np.float32)
y_train_np = (2*X_train_np[:, 0] + 3*X_train_np[:, 1] + 
              np.random.randn(500).astype(np.float32) * 0.5)

# Convert to PyTorch tensors
X_train_torch = torch.from_numpy(X_train_np)
y_train_torch = torch.from_numpy(y_train_np).reshape(-1, 1)

# Create dataset and dataloader
dataset = TensorDataset(X_train_torch, y_train_torch)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Build model
model_torch = nn.Sequential(
    nn.Linear(2, 10),
    nn.ReLU(),
    nn.Linear(10, 5),
    nn.ReLU(),
    nn.Linear(5, 1)
)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model_torch.parameters(), lr=0.01)

print("Model Architecture:")
print(model_torch)
print()

# Training loop
print("Training...")
losses_torch = []
epochs = 100

for epoch in range(epochs):
    epoch_loss = 0
    for X_batch, y_batch in dataloader:
        # Forward pass
        predictions = model_torch(X_batch)
        loss = criterion(predictions, y_batch)
        
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
    
    avg_loss = epoch_loss / len(dataloader)
    losses_torch.append(avg_loss)
    
    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.6f}")

# Plot training loss
plt.figure(figsize=(10, 6))
plt.plot(losses_torch, linewidth=2, color='green')
plt.title('Training Loss (PyTorch)', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.grid(True, alpha=0.3)
plt.show()

print("\n✓ PyTorch model trained successfully!")
```

```python
# Classification example with PyTorch
print("=== PyTorch Classification Example ===")
print()

# Use same dataset
X_train_clf = torch.from_numpy(X_train.astype(np.float32))
y_train_clf = torch.from_numpy(y_train.astype(np.float32)).reshape(-1, 1)
X_test_clf = torch.from_numpy(X_test.astype(np.float32))
y_test_clf = torch.from_numpy(y_test.astype(np.float32)).reshape(-1, 1)

# Build model
class ClassificationNet(nn.Module):
    def __init__(self):
        super(ClassificationNet, self).__init__()
        self.fc1 = nn.Linear(2, 16)
        self.fc2 = nn.Linear(16, 8)
        self.fc3 = nn.Linear(8, 1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.sigmoid(self.fc3(x))
        return x

clf_model_torch = ClassificationNet()
criterion_clf = nn.BCELoss()
optimizer_clf = optim.Adam(clf_model_torch.parameters(), lr=0.01)

# Create dataloader
train_dataset_clf = TensorDataset(X_train_clf, y_train_clf)
train_loader_clf = DataLoader(train_dataset_clf, batch_size=32, shuffle=True)

# Training
for epoch in range(50):
    for X_batch, y_batch in train_loader_clf:
        predictions = clf_model_torch(X_batch)
        loss = criterion_clf(predictions, y_batch)
        
        optimizer_clf.zero_grad()
        loss.backward()
        optimizer_clf.step()

# Evaluate
clf_model_torch.eval()
with torch.no_grad():
    test_preds = clf_model_torch(X_test_clf)
    test_preds_binary = (test_preds > 0.5).float()
    accuracy = (test_preds_binary == y_test_clf).float().mean()

print(f"Test Accuracy: {accuracy:.4f}")
print("\n✓ PyTorch classification model trained!")
```

## 10. Section 9: CNN Example (Image Classification)

Let's build a Convolutional Neural Network for image classification:

```python
# Load MNIST dataset
print("=== CNN for Image Classification (MNIST) ===")
print()

(X_train_img, y_train_img), (X_test_img, y_test_img) = keras.datasets.mnist.load_data()

# Normalize and reshape
X_train_img = X_train_img.reshape(-1, 28, 28, 1).astype('float32') / 255.0
X_test_img = X_test_img.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# One-hot encode labels
y_train_img = keras.utils.to_categorical(y_train_img, 10)
y_test_img = keras.utils.to_categorical(y_test_img, 10)

print(f"Training data shape: {X_train_img.shape}")
print(f"Test data shape: {X_test_img.shape}")
print()

# Build CNN
cnn_model = keras.Sequential([
    # First convolutional block
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1), name='conv1'),
    layers.MaxPooling2D((2, 2), name='pool1'),
    
    # Second convolutional block
    layers.Conv2D(64, (3, 3), activation='relu', name='conv2'),
    layers.MaxPooling2D((2, 2), name='pool2'),
    
    # Flatten and dense layers
    layers.Flatten(name='flatten'),
    layers.Dense(128, activation='relu', name='dense1'),
    layers.Dropout(0.5, name='dropout'),
    layers.Dense(10, activation='softmax', name='output')
])

cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("CNN Architecture:")
cnn_model.summary()
print()
```

```python
# Train CNN
print("Training CNN...")
cnn_history = cnn_model.fit(
    X_train_img, y_train_img,
    epochs=5,
    batch_size=128,
    validation_data=(X_test_img, y_test_img),
    verbose=1
)

# Evaluate
test_loss, test_acc = cnn_model.evaluate(X_test_img, y_test_img, verbose=0)
print(f"\nTest Accuracy: {test_acc:.4f}")

# Plot results
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(cnn_history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(cnn_history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('CNN Loss', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(cnn_history.history['accuracy'], label='Training Accuracy', linewidth=2)
plt.plot(cnn_history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.title('CNN Accuracy', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n✓ CNN trained successfully!")
```

## 11. Section 10: RNN Example (Sequence Prediction)

Let's build a simple RNN for sequence prediction:

```python
# Generate synthetic sequence data
print("=== RNN for Sequence Prediction ===")
print()

def generate_sequence_data(n_samples=1000, seq_length=10):
    """
    Generate sequences where the target is the sum of the sequence.
    """
    X = np.random.randn(n_samples, seq_length, 1)
    y = np.sum(X, axis=1)
    return X.astype(np.float32), y.astype(np.float32)

X_seq, y_seq = generate_sequence_data()
print(f"Sequence data shape: {X_seq.shape}")
print(f"Target shape: {y_seq.shape}")
print()

# Build RNN model
rnn_model = keras.Sequential([
    layers.LSTM(32, return_sequences=True, input_shape=(10, 1), name='lstm1'),
    layers.LSTM(16, name='lstm2'),
    layers.Dense(1, name='output')
])

rnn_model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

print("RNN Architecture:")
rnn_model.summary()
print()
```

```python
# Train RNN
print("Training RNN...")
rnn_history = rnn_model.fit(
    X_seq, y_seq,
    epochs=20,
    batch_size=32,
    validation_split=0.2,
    verbose=0
)

# Plot training history
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(rnn_history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(rnn_history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('RNN Loss', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(rnn_history.history['mae'], label='Training MAE', linewidth=2)
plt.plot(rnn_history.history['val_mae'], label='Validation MAE', linewidth=2)
plt.title('RNN MAE', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n✓ RNN trained successfully!")
```

```python
# Test RNN predictions
X_test_seq, y_test_seq = generate_sequence_data(n_samples=10)
predictions_seq = rnn_model.predict(X_test_seq, verbose=0)

print("=== RNN Predictions ===")
print()
for i in range(5):
    print(f"Sequence {i+1}:")
    print(f"  True sum:      {y_test_seq[i][0]:.4f}")
    print(f"  Predicted sum: {predictions_seq[i][0]:.4f}")
    print(f"  Error:         {abs(y_test_seq[i][0] - predictions_seq[i][0]):.4f}")
    print()

print("✓ RNN predictions demonstrated!")
```

### Project 2: Compare All Frameworks on Same Task

```python
print("=== Project 2: Framework Comparison ===")
print()
print("Task: Binary classification on synthetic dataset")
print()

# Generate dataset
X_comp, y_comp = make_moons(n_samples=1000, noise=0.2, random_state=42)
X_train_comp, X_test_comp, y_train_comp, y_test_comp = train_test_split(
    X_comp, y_comp, test_size=0.2, random_state=42
)

results = {}

# 1. NumPy implementation
print("Training NumPy model...")
numpy_model = NeuralNetwork([2, 16, 8, 1], learning_rate=0.01)
numpy_model.train(X_train_comp, y_train_comp, epochs=100, batch_size=32, verbose=False)
numpy_preds = (numpy_model.predict(X_test_comp) > 0.5).astype(int).flatten()
results['NumPy'] = (numpy_preds == y_test_comp).mean()

# 2. TensorFlow implementation
print("Training TensorFlow model...")
tf_model = keras.Sequential([
    layers.Dense(16, activation='relu', input_shape=(2,)),
    layers.Dense(8, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
tf_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
tf_model.fit(X_train_comp, y_train_comp, epochs=50, batch_size=32, verbose=0)
_, tf_acc = tf_model.evaluate(X_test_comp, y_test_comp, verbose=0)
results['TensorFlow'] = tf_acc

# 3. PyTorch implementation
print("Training PyTorch model...")
X_train_pt = torch.from_numpy(X_train_comp.astype(np.float32))
y_train_pt = torch.from_numpy(y_train_comp.astype(np.float32)).reshape(-1, 1)
X_test_pt = torch.from_numpy(X_test_comp.astype(np.float32))
y_test_pt = torch.from_numpy(y_test_comp.astype(np.float32)).reshape(-1, 1)

pt_model = ClassificationNet()
pt_criterion = nn.BCELoss()
pt_optimizer = optim.Adam(pt_model.parameters(), lr=0.01)
pt_dataset = TensorDataset(X_train_pt, y_train_pt)
pt_loader = DataLoader(pt_dataset, batch_size=32, shuffle=True)

for epoch in range(50):
    for X_batch, y_batch in pt_loader:
        pt_preds = pt_model(X_batch)
        loss = pt_criterion(pt_preds, y_batch)
        pt_optimizer.zero_grad()
        loss.backward()
        pt_optimizer.step()

pt_model.eval()
with torch.no_grad():
    pt_test_preds = pt_model(X_test_pt)
    pt_test_preds_binary = (pt_test_preds > 0.5).float()
    pt_acc = (pt_test_preds_binary == y_test_pt).float().mean().item()
results['PyTorch'] = pt_acc

# Display results
print("\n" + "="*50)
print("RESULTS")
print("="*50)
for framework, accuracy in results.items():
    print(f"{framework:15s}: {accuracy:.4f} ({accuracy*100:.2f}%)")
print("="*50)

# Visualize comparison
plt.figure(figsize=(10, 6))
frameworks = list(results.keys())
accuracies = list(results.values())
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
bars = plt.bar(frameworks, accuracies, color=colors, edgecolor='black', linewidth=2)

plt.title('Framework Comparison: Binary Classification Accuracy', 
          fontsize=16, fontweight='bold')
plt.ylabel('Accuracy', fontsize=12)
plt.ylim(0.85, 1.0)
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar, acc in zip(bars, accuracies):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{acc:.4f}',
             ha='center', va='bottom', fontweight='bold', fontsize=12)

plt.tight_layout()
plt.show()

print("\n✓ Framework comparison complete!")
```
